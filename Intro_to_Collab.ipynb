## 🛠️ The Tools of the Trade
To build AI, we don't code everything from scratch. We use **Libraries** (sets of pre-made tools). Today we will use:
* **NumPy:** For handling numbers and data.
* **Matplotlib:** For creating cool charts and graphs.

```python
import matplotlib.pyplot as plt
import numpy as np

# Let's create some data points
x = np.linspace(0, 10, 100)
y = np.sin(x) # A simple wave pattern

# Let's visualize it
plt.figure(figsize=(8,4))
plt.plot(x, y, color='teal', linewidth=2, label='Sine Wave')
plt.title("Our First Visualization", fontsize=14)
plt.xlabel("Time")
plt.ylabel("Intensity")
plt.grid(True, linestyle='--')
plt.legend()
plt.show()
```
